## Identifying missing values in the dataset:
In order to keep a better overview, we will visualize missing values per columns. Objective is to plot the variables with more and less than 10% missing values.

```python
def missing_data(data, thresh=1, color='black', edgecolor='black', width=15, height=3):
     
    plt.figure(figsize=(width,height))
    percentage=(data.isnull().mean())*100
    percentage.sort_values(ascending=False)[:10].plot.bar(color=color, edgecolor=edgecolor)
    plt.axhline(y=thresh, color='r', linestyle='-')
    plt.title('Missing values percentage per column', fontsize=20, weight='bold' )
    plt.text(len(data.isnull()[:10].sum()/len(data))/15, thresh +8, 
             f'Columns with more than {thresh}% missing values', fontsize=12, color='darkblue',
             ha='left' ,va='top')
    plt.text(len(data.isnull()[:10].sum()/len(data))/15, thresh -3, 
             f'Columns with less than {thresh}% missing values', fontsize=12, color='green',
             ha='left' ,va='top')
    plt.xlabel('Columns', size=15, weight='bold')
    plt.ylabel('Missing values percentage')
    
    return plt.show()
```

###A) K-means Clustering


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
from sklearn.preprocessing import StandardScaler
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer

# Define a function for K-means clustering
def kmeans_clustering(df_scaled, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=100, random_state=42)
    kmeans.fit(df_scaled)
    labels = kmeans.labels_
    inertia = kmeans.inertia_
    sil_score = silhouette_score(df_scaled, labels)
    return labels, inertia, sil_score

# Define the range of clusters to try
clusters = range(2, 11)

# Create a dictionary to store the results
results = {'labels': {}, 'inertia': {}, 'sil_score': {}}

# Iterate over the cluster range and perform K-means clustering
for n_clusters in clusters:
    labels, inertia, sil_score = kmeans_clustering(df_scaled, n_clusters)
    results['labels'][n_clusters] = labels
    results['inertia'][n_clusters] = inertia
    results['sil_score'][n_clusters] = sil_score

# Plot the elbow method for inertia
plt.figure(figsize=(12, 6))
plt.plot(list(results['inertia'].keys()), list(results['inertia'].values()), linewidth=2, color='green', marker='8')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow method for Inertia', fontsize=18)
plt.show()

# Plot the elbow method for silhouette score
plt.figure(figsize=(12, 6))
plt.plot(list(results['sil_score'].keys()), list(results['sil_score'].values()), linewidth=2, color='brown', marker='8')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Elbow method for Silhouette Score', fontsize=18)
plt.show()

# Select the best number of clusters based on the elbow methods
best_n_clusters = 3  # replace with the elbow method result

# Perform K-means clustering with the best number of clusters
labels, inertia, sil_score = kmeans_clustering(df_scaled, best_n_clusters)

# Print the results
print(f'The clusters are: {labels}')
print(f'The inertia is: {inertia}')
print(f'The silhouette score is: {sil_score}')

# Plot the clusters
visualizer = SilhouetteVisualizer(KMeans(n_clusters=best_n_clusters, init='k-means++', n_init=10, max_iter=100, random_state=42), colors='BrBG')
visualizer.fit(df_scaled)
visualizer.show()

# Create a dataframe of the cluster counts
unique, counts = np.unique(labels, return_counts=True)
```

In the improved code, a function was used for repetitive tasks to reduce the code and make it more manageable. A loop was used to iterate over different numbers of clusters instead of hardcoding the number of clusters, which made it easy to change the range without having to change the code. A dictionary was used to store the results instead of creating separate variables for each result, making it easier to manage and access the results.

E) Agglomerative Hierarchical Clustering

```python
import scipy.cluster.hierarchy as shc
import sklearn.metrics as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score


def plot_dendrogram(df_pca):
    plt.figure(figsize=(20, 10))
    plt.title("Dendrograms")
    dend = shc.dendrogram(shc.linkage(df_pca, method='complete'))
    plt.axhline(y=0.8, color='r', linestyle='--')
    plt.show()


def hierarchical_clustering(df_pca, n_clusters, linkage_type):
    hc = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage=linkage_type)
    hc_model = hc.fit(df_pca)
    hc_labels = hc_model.labels_
    silh_score = silhouette_score(df_pca, hc_labels)
    print(f'Silhouette Coefficient for {linkage_type} linkage with {n_clusters} clusters:', silh_score)
    return hc_labels


def plot_clusters(df_pca_2, labels, title):
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=df_pca_2, x='Component 1', y='Component 2', hue=labels, palette='BrBG')
    plt.title(title, fontsize=18)
    plt.show()


if __name__ == '__main__':
    # Step 1: Data pre-processing (already completed in earlier steps)
    
    # Step 2: Finding the optimal number of clusters using the dendrogram
    plot_dendrogram(df_pca)
    
    # Step 3: Training the hierarchical clustering model
    n_clusters = 8
    hc_ward_labels = hierarchical_clustering(df_pca, n_clusters, 'ward')
    df_pca_2['hc_ward_labels'] = hc_ward_labels

    # Step 4: Visualization of the clusters
    plot_clusters(df_pca_2, hc_ward_labels, 'Hierarchical "Ward" linkage')
```
