```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/drive/MyDrive/146/finL PROJECT/Data_Cortex_Nuclear.csv')
df = data.copy()
pd.set_option('display.max_row',df.shape[0])
pd.set_option('display.max_column',df.shape[1]) 
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.isna(),cbar=False)
plt.show()
```

```python
df.shape
```

```python
df1 = df.dropna(axis=1, thresh=df.shape[0]*0.9)
```

```python
df1.shape
```

```python
df1.isnull().sum().sort_values(ascending=False)
```

```python
df1['pCFOS_N'].describe()
```

```python
df2 = df1.fillna(df1.mean())
```

```python
duplicateRowsdf = df2[df2.duplicated()]
print(duplicateRowsdf)
```

```python
df2.describe()
```

```python
df2.describe(include='object')
```

```python
df2 = df2.drop('MouseID', axis=1)
```

```python
print(df2['Genotype'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')  # Set the style of the plot

# Create the count plot
ax = sns.countplot(x='Genotype', data=df2, palette='YlOrBr')

# Set the title and axis labels
ax.set_title('Distribution of Genotype', fontsize=18, pad=20)
ax.set_xlabel('Genotype', fontsize=14)
ax.set_ylabel('Count', fontsize=14)

# Customize tick labels and rotation if needed
ax.tick_params(axis='x', labelrotation=45)

# Add value labels on top of each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 5),
                textcoords = 'offset points',
                fontsize=10,
                color='black')

# Adjust the plot spacing
plt.tight_layout()

# Show the plot
plt.show()
```
